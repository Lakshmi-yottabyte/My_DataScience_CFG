```python
import requests
```

```python
response = requests.get('http://127.0.0.1:5000')
print(response.text)
```

```python
response = requests.get('http://127.0.0.1:5000/hello?name=Alan')

print(response.text)
```

```python
# add new user

new_user = {'name': 'Alan', 'age': 39}

response = requests.post('http://127.0.0.1:5000/add_user', json=new_user)

print(response.status_code)

print(response.text)
```
